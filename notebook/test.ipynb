```python
import sys
sys.path.append('../src/')
from ActivationFunction import *
import PlotUtils
import pandas as pd
import numpy as np
pd.options.plotting.backend = "plotly"

generator = np.random.default_rng()


AF = {'sigmoid' : SigmoidFunction(),
      'identity' : IdentityFunction(),
      'tanh' : HyperbolicTangentFunction(1),
      'ReLU' : RectifiedLinearUnit(),
      'ELU' : ExponentialLinearUnit()}

data = {}
xs = np.arange(-5, 5, 0.01)
for fun in AF:
    
    data[fun] = AF[fun].compute(xs)
    data[fun + '_deriv'] = AF[fun].derivate(xs)

pd.DataFrame(data).plot()
```
